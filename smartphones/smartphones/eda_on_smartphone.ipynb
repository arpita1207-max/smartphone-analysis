```python
num_cols=df.iloc[:,[2,3,9,12,14,15,16,23,24]].columns
```

```python
# univariate analysis for numerical columns
def plot_univariate_graphs(df,columns):
  for col in columns:
    fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(7,4))
    ax=ax.flatten()
    # to check frequency
    print(f'Skewness of {col} : {df[col].skew()}')
    sns.histplot(df[col],ax=ax[0],kde=True,label=col)
    ax[0].set_xlabel(col)
    ax[0].set_ylabel('Frequency')
    ax[0].set_title(f'Hist Plot of {col}')
    #to check outliers
    sns.boxplot(df[col],ax=ax[1])
    ax[1].set_xlabel(f'{col}')
    ax[1].set_ylabel('Frequency')
    ax[1].set_title(f'Box Plot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
plot_univariate_graphs(df,num_cols)
```

#####  Price:-Fairly right skewed(phones with higher prices are less).Outliers are there.
#####  Rating:- Moderately normal distribution (most ratings are between 75 and 85).No Outliers
#####  Processor Speed:-Bimodal Distribution(Processor Speed are between 2 & 3).No Outliers
#####  Fast Charging:-Multimodal Distribution(some phones have fast charging of 140 and above).Outliers are there.
#####  Screen Size:-fairly left skewed distribution(some phones have screen size<6 and below and some have >7 and above).Outliers are there
#####  ppi:-Multimodal Distribution(some phones have less than 300 and below ppi).Outliers are there.
#####  Primary Rear Camera:-Multimodal Distribution(1 phone have more than 200MP camera).Outlier is there.
#####  Primary Front Camera:-Multimodal Distribution(some phones have front camera of 30 and below).Outliers are there.
#####  Battery Capacity:-Multimodal Distribution(some phones have battery capacity of 6000 and above)).Outliers are there.



```python
# bivariate analysis for numerical columns
def plot_bivariate_graphs_num(columns):
    for col in columns:
      fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(5,4))
      # to relationship of both columns
      sns.scatterplot(data=df,x=df[col],y=df['price'],ax=ax)
      ax.set_xlabel(col)
      ax.set_ylabel('Price')
      ax.set_title(f'Scatter Plot of {col} vs price')
      plt.xticks(rotation=90)
      plt.tight_layout()
      plt.show()

```

```python
num_cols=df.iloc[:,[3,9,12,14,15,16,23,24]].columns
plot_bivariate_graphs_num(num_cols)
```

```python
# bivariate analysis for categorical columns
def plot_bivariate_graphs_cat(columns):
      for i in columns:
        plt.figure(figsize=(7,5))
        sns.barplot(data=df,x=df[i],y=df['price'],estimator=np.median,width=0.5,palette='viridis')
        plt.xlabel(i)
        plt.ylabel('Price')
        plt.xticks(rotation=90)
        plt.title(f'Bar Plot of {i} vs price')
        plt.show()

```

```python
cat_cols=(df.columns[[0,4,5,6,7,8,10,11,13,17,18,19,20,21,22,25]])
cat_cols
```

```python
plot_bivariate_graphs_cat(cat_cols)
```

```python
from sklearn.impute import KNNImputer
knn=KNNImputer(n_neighbors=5, weights='uniform',missing_values=np.nan)
impute_cols=['rating', 'processor_speed', 'battery_capacity',
       'extended_upto_memory', 'primary_front_camera']
xformed_df=pd.DataFrame(knn.fit_transform(df[impute_cols]),columns=impute_cols)
new_df=pd.concat([df.drop(columns=impute_cols),xformed_df],axis=1)
```

```python
new_df['processor_brand']=new_df['processor_brand'].fillna(new_df['processor_brand'].mode()[0])
```

```python

corr_df=pd.DataFrame(df[df.select_dtypes(exclude=['object']).columns].corr()['price']).rename(columns={'price':'correlation coefficient old'})
```

```python
corr_new_df=pd.DataFrame(new_df[new_df.select_dtypes(exclude=['object']).columns].corr()['price']).rename(columns={'price':'correlation coefficient new'})
```
