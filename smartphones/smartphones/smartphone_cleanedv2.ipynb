```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('/content/smartphone-cleaned.csv')
```

```python
df.shape
```

```python
def fast_charging(text):
  if 'Fast Charging' in text:
    return 1
  return 0
```

```python
fast_charging_available=df['battery'].replace(np.nan,'').apply(fast_charging)
```

```python
df.insert(loc=19,column='fast_charging_available',value=fast_charging_available)
#df.drop(columns='fast_charging_available',inplace=True)
```

```python
df['fast_charging_available'].value_counts(dropna=False)
```

```python
def os_category(text):
  if 'Android' in text:
    return 'android'
  elif 'iOS' in text:
    return 'ios'
  else:
    return 'others'
```

```python
df['os']=df['os'].replace(np.nan,'None').apply(os_category)
```

```python
df['os'].value_counts(dropna=False)
```

```python

primary_rear_camera=df['camera'].str.strip().str.split('&').str.get(0).str.split('+').str.get(0).str.findall(r'^\d+\.?\d*').str.get(0).str.strip().astype('float')
```

```python
df.insert(loc=29,column='primary_rear_camera',value=primary_rear_camera)
#df.drop(columns='primary_rear_camera',inplace=True)
```

```python
primary_front_camera=df['camera'].str.strip().str.split('&').str.get(1).str.findall(r'\d+').str.get(0).str.strip().astype('float')
```

```python
df.insert(loc=30,column='primary_front_camera',value=primary_front_camera)
#df.drop(columns='primary_front_camera',inplace=True)
```

```python
def extended_memory(text):
    if 'TB' in text:
      return str(int(text.split('upto')[1].replace('TB',''))*1024)
    elif 'GB' in text:
      return str(text.split('upto')[1].replace('GB',''))
    elif 'Not' in text:
      return np.nan
    else:
      return np.nan
    return np.nan


```

```python
extended_upto_memory=df['card'].apply(extended_memory).astype('float')
```

```python
df.insert(loc=29,column='extended_upto_memory',value=extended_upto_memory)
```

```python
def ext_memory_available(text):
  if 'Not' in text:
    return 0
  else:
    return 1
```

```python
extended_memory_available=df['card'].apply(ext_memory_available)
```

```python
df.insert(loc=30,column='extended_memory_available',value=extended_memory_available)
```

```python
df['num_cores'].value_counts(dropna=False)
```

```python
df.loc[[152, 262, 280, 444, 706],'num_cores']='Hexa Core'
df.loc[425,'num_cores']='Octa Core'
```

```python
def convert_core_to_num(text):
  if 'Quad' in text:
    return 4.0
  elif 'Octa' in text:
    return 8.0
  elif 'Hexa' in text:
    return 6.0
  elif 'Dual' in text:
    return 2.0
  else:
    return 1.0
```

```python
df['num_cores']=df['num_cores'].apply(convert_core_to_num)
```

```python
df['num_cores'].value_counts()
```

```python
width=df['resolution'].str.strip().str.split('x').str.get(0).str.strip().astype('float')
height=df['resolution'].str.strip().str.split('x').str.get(1).str.strip().astype('float')
```

```python
df.insert(loc=17,column='width',value=width)
df.insert(loc=18,column='height',value=height)
```

```python
ppi=np.round(np.sqrt((df['width']*df['width'])+(df['height']*df['height']))/(df['screen_size']),2)
df.insert(loc=19,column='ppi',value=ppi)
```

```python
def ppi_range(x):
  if x<=300:
    return 'Low PPI'
  elif x>300 or x<=450:
    return 'Medium PPI'
  else:
    return 'High PPI'
```

```python
pixel_range=df['ppi'].apply(ppi_range)
df.insert(loc=20,column='pixel_range',value=pixel_range)
```

```python
df.drop(columns=['width','height','display','resolution'],inplace=True)
```

```python
df['processor_brand'] =df['processor_brand'].replace('Sanpdragon','Snapdragon').replace('Apple','Bionic').str.strip()
```
