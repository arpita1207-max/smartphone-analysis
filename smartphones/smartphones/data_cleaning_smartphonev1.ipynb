```python
def fast_charging_extractor(item):
  if type(item)==list:
     if len(item)==1:
      return item[0]
     else:
      return 0
  else:
    return -1

```

```python
def camera_extractor(text):
  if 'Triple' in text:
    return '3'
  elif 'Dual' in text:
    return '2'
  elif 'Quad' in text:
    return '4'
  elif 'Missing' in text:
    return 'Missing'
  else:
    return '1'
```

```python
num_rear_cameras=df['camera'].str.strip().str.split('&').str.get(0).fillna('Missing').apply(camera_extractor).str.strip()
```

```python
num_front_cameras=df['camera'].str.strip().str.split('&').str.get(1).fillna('Missing').apply(camera_extractor).str.strip()
```

```python
df.insert(loc=25,column='num_rear_cameras',  value=num_rear_cameras.values)
df.insert(loc=26,column='num_front_cameras',  value=num_front_cameras.values)
#df.drop(columns=['num_front_cameras','num_rear_cameras'],inplace=True)
```
